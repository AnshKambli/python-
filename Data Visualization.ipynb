```python
# create a data visualizaion using matplotlib
```

```python
import matplotlib.pyplot as plt
a=(25,45,23,45,60)
b=('a','b','c','d','e')
plt.bar(b,a)
plt.title('bar chart')
plt.show()
plt.pie(a,labels=b)
plt.title('pie chart')
plt.show()
plt.scatter(a,b)
plt.title('scatter chart')
plt.show()
plt.hist(a)
plt.title('histogram chart')
plt.show()
plt.boxplot(a)
plt.title('boxplot chart')
plt.show()
```
